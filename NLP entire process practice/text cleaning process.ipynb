# Input

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP & LLMs/NLP entire process practice/data source/Copy of pdata.csv')

data.head()
```

```python
#text_data = data['citationStringAnnotated'][:500].copy()
text_data = data['citationStringAnnotated'].copy()

text_data.head()
```

# Phase 1: String-Level Cleaning

```python
text = text_data.copy()

text.head()
```

## Text lowercasing:

```python
lowercased_text= text.apply(lambda x: x.lower())

lowercased_text.head()
```

```python
text = lowercased_text.copy()
```

## Handling Contractions:

```python
!pip install contractions
```

```python
import contractions

# This automatically finds and expands virtually every English contraction
contractions_text = text.apply(lambda txt: contractions.fix(str(txt)))

contractions_text.head()
```

```python
text = contractions_text.copy()
```

## Handling Special characters:

```python
import re

special_characters_removed_text = text.str.replace(r'[^\w\s]', ' ', regex=True)

special_characters_removed_text.head()
```

```python
text = special_characters_removed_text.copy()
```

## Handling Numerical values:

```python
import re

digit_removed_text = text.str.replace(r'\d+', '', regex=True)

digit_removed_text.head()
```

```python
text = digit_removed_text.copy()
```

## Whitespace stripping:

```python
text_stripped= text.apply(lambda txt: txt.strip())

text_stripped.head()
```

```python
text = text_stripped.copy()
```

# Phase 2: Tokenization

## Tokenization (a technique for NLP preprocessing):

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the required tokenizer models (only needs to be run once)
nltk.download('punkt_tab', quiet=True)

# Apply tokenization to every row in your dataset
# Note: This changes your column from Strings into Lists of Strings!
text_tokens = text.apply(lambda txt: word_tokenize(str(txt)))

text_tokens.head()
```

```python
text = text_tokens.copy()
```

# Phase 3: Word-Level Processing

## Stop-word removal:

```python
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))

# Use .apply() to run the list comprehension on every row
stop_word_removed_text = text.apply(lambda txt: ' '.join([word for word in str(txt).split() if word.lower() not in stop_words]))

display(stop_word_removed_text.head())
```

```python
text = stop_word_removed_text.copy()
```

## Lemmatization/ Stemming:

```python
import nltk
nltk.download('wordnet', quiet=True)
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = text.apply(lambda txt: ' '.join([lemmatizer.lemmatize(word) for word in str(txt).split()]))

lemmatized_tokens.head()
```

```python
text = lemmatized_tokens.copy()
```

## Removal of short/ rare words:

```python
from collections import Counter

# 1. Combine ALL rows into one massive string, then split into a list of all words
all_words = ' '.join(text.astype(str)).split()

# 2. Count the frequencies of every word in the entire dataset
corpus_freq = Counter(all_words)

# 3. Create a fast-lookup set of words we want to KEEP (appearing 2 or more times)
threshold = 2
words_to_keep = set(word for word, count in corpus_freq.items() if count >= threshold)

# 4. Apply the filter to every row in the dataframe
rare_words_removed_text = text.apply(lambda txt: ' '.join([word for word in str(txt).split() if word in words_to_keep]))

rare_words_removed_text.head()
```

```python
text = rare_words_removed_text.copy()
```
